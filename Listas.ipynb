```python
# Programa para armazenar nomes de alunos em uma lista

alunos = []  # Lista vazia para guardar os nomes

print("Cadastro de Alunos ")
print("Digite 'ok' para encerrar.\n")

while True:
    nome = input("Digite o nome do aluno: ")

    if nome.lower() == "ok":   # Verifica se o usuário quer encerrar
        break

    alunos.append(nome)  # Adiciona o nome à lista

print("\n = Lista de Alunos Cadastrados ")
for aluno in alunos:
    print(aluno)
```

O programa começa criando uma lista vazia chamada alunos. Essa lista funciona como uma caixinha onde iremos guardar todos os nomes digitados. Em seguida, o programa mostra uma mensagem explicando que é um cadastro de alunos e avisa que, se o usuário digitar “ok”, o processo termina. Depois disso, o código entra em um loop while True, que basicamente significa “continue rodando sem parar até eu mandar parar”. Dentro desse estrutura, o programa pede ao usuário que digite o nome de um aluno. Se a pessoa digitar “ok”, usando letras maiúsculas ou minúsculas, o programa reconhece isso por causa do .lower() e usa o comando break para sair do ciclo de repetição. Se não for “sair”, o nome digitado é colocado dentro da lista usando alunos.append(nome). Quando o usuário decide parar e digita “ok”, a repetição acaba e o programa mostra um título indicando que agora vai exibir todos os alunos cadastrados.
